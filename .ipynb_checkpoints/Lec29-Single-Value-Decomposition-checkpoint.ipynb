# Single Value Decomposition (SVD)

$$A = U\Sigma V^T$$

$\Sigma$ is diagonal, $U, V$ is orthogonal

Family of matrix studied before is symmetric positive definite

$$A = S\Lambda S^{-1}$$

Usually eigenvector matrix is not orthogonal (Not good). But in symmetric case, the eigenvector matrix is orthogonal and its inverse is same as its transpose. And because it is positive definite, the ordinary eigenvalues become positive eigenvalues

$$A = Q\Lambda Q^T$$

Suppose in row space ($R^n$), we have $v_1$ and in column space ($R^m$), we have $\sigma_1 u_1 = Av_1$. $v_1$ is an unit vector, and $u_1$ need not to be length 1, so $\sigma_1$ is the stretching factor

We want to have a vector $v_2$ which is orthogonal to $v_1$, and also goes into orthogonal basis in column space, $\sigma_2 u_2 = Av_2$. How to find such $v_2$?

Ino the words, we want to achieve this

$$A \begin{bmatrix}\vec v_1 & \vec v_2 & ... & \vec v_r \end{bmatrix} =
\begin{bmatrix}\vec u_1 & \vec u_2 & ... & \vec u_r \end{bmatrix}
\begin{bmatrix}\sigma_1 &  & \\  & \sigma_2 &  \\ & & \ddots \\ & & & \sigma_r \end{bmatrix}$$

$$AV = U\Sigma \iff A = U\Sigma V^T$$

$V$ is orthonormal basis in row space, $U$ is orthonormal basis in column space, $\Sigma$ is a diagonalize matrix

### Example

$$A = \begin{bmatrix}4 & 4 \\ -3 & 3 \end{bmatrix}$$

We want to find $v_1, v_2$ (orthonormal) in row space $R^2$

$u_1, u_2$ (orthonormal) in column space $R^2$

$\sigma_1 > 0, \sigma_2 > 0$

$Av_1 = \sigma_1 u_1 \quad Av_2 = \sigma_2 u_2$

Because $A = U\Sigma V^T$

$$A^TA = V\Sigma^T U^T U\Sigma V^T = V\Sigma^2 V^T$$

So $U$ disappear in the equation

$$A^T A = \begin{bmatrix}25 & 7 \\ 7 & 25 \end{bmatrix}$$

Its eigenvectors will be $V$, and eigenvalues is the $\Sigma^2$

$A^TA \begin{bmatrix}1 \\ 1 \end{bmatrix} = 32 \begin{bmatrix}1 \\ 1 \end{bmatrix}$

$A^TA \begin{bmatrix}1 \\ -1 \end{bmatrix} = 18 \begin{bmatrix}1 \\ -1 \end{bmatrix}$

Normalize the eigenvectors, $v_1 = \frac{1}{\sqrt{2}}\begin{bmatrix}1 \\ 1 \end{bmatrix}, v_2 = \frac{1}{\sqrt{2}}\begin{bmatrix}1 \\ -1 \end{bmatrix}$

$$A = U\Sigma V^T$$

$$\begin{bmatrix}4 & 4 \\ -3 & 3 \end{bmatrix} = \begin{bmatrix} & \\ &\end{bmatrix} \begin{bmatrix}\sqrt{32} & 0 \\ 0 & \sqrt{18} \end{bmatrix} \frac{1}{\sqrt{2}} \begin{bmatrix}1 & 1 \\ 1 & -1 \end{bmatrix}$$

Now find $U$

$$AA^T = U\Sigma V^T V\Sigma^T U^T = U\Sigma^2 U^T$$

$$AA^T = \begin{bmatrix}32 & 0 \\ 0 & 18 \end{bmatrix}$$

This is an accident where the result comes out diagonal, so we know eigenvalues is 32 and 18. And the eigenvectors are $\begin{bmatrix}1 \\ 0 \end{bmatrix}, \begin{bmatrix}0 \\ 1 \end{bmatrix}$

But eigenvalues 32 and 18 are not accident. Eigenvalues $A^T A$ are equal to $AA^T$

So, we found $U$

$$\begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix} \begin{bmatrix}\sqrt{32} & 0 \\ 0 & \sqrt{18} \end{bmatrix} \frac{1}{\sqrt{2}} \begin{bmatrix}1 & 1 \\ 1 & -1 \end{bmatrix} = \begin{bmatrix}4 & 4 \\ 3 & -3 \end{bmatrix} \neq \begin{bmatrix}4 & 4 \\ -3 & 3 \end{bmatrix}$$

Why?




## Example 2

$$A = \begin{bmatrix}4 & 3 \\ 8 & 6 \end{bmatrix}$$

Rank 1 matrix, singular matrix. Row space is a multiple of $\begin{bmatrix}4 \\ 3 \end{bmatrix}$ and $\bot N(A)$. Column space is multiple of $\begin{bmatrix}4 \\ 8 \end{bmatrix}$ and $\bot N(A^T)$

Make $v_1$ unit vector, $v_1 = \begin{bmatrix}.8 \\ .6 \end{bmatrix}$. $v_2 \bot v_1$, so $v_2 = \begin{bmatrix}.6 \\ .8 \end{bmatrix}$

Make $u_1$ unit vector, $u_1 = \frac{1}{\sqrt{5}}\begin{bmatrix}1 \\ 2 \end{bmatrix}$

$$A^T A = \begin{bmatrix}80 & 60 \\ 60 & 45 \end{bmatrix}$$

eigenvalues are 0, 125

$$A = U\Sigma V^T$$

$$\begin{bmatrix}4 & 3 \\ 8 & 6 \end{bmatrix} = 
\frac{1}{\sqrt{5}} \begin{bmatrix}1 & 2 \\ 2 & -1 \end{bmatrix}
\begin{bmatrix}\sqrt{125} & 0 \\ 0 & 0 \end{bmatrix}
\begin{bmatrix}.8 & .6 \\ .6 & -.8 \end{bmatrix}$$

We are choosing the right basis for the four subspaces

$v_1, ..., v_r$ is the orthonormal basis for row space

$u_1, ..., u_r$ is the orthonormal basis for column space

$v_{r+1}, ..., v_n$ is the orthonormal basis for the nullspace

$u_{r+1}, ..., u_m$ is the orthonormal basis for the nullspace of A transpose

We needed eigenvalues because these basis make A diagonal. $Av_i = \sigma_i u_i$. When we choose these basis, there is no coupling between v's and no coupling between u's. Each Av is in each direction of u
