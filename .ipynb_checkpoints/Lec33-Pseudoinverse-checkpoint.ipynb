# Left and Right Inverse

2-sided inverse

$AA^{-1} = I = A^{-1}A$$

This is true when r = m = n, full rank

## Left Inverse

If there is left inverse, then there is full column rank r = n (columns all independent but rows not necessarily and number of rows might be less than number of columns)

Nullspace is zero vector because m independent columns. So we have 0 or 1 solution to $Ax = b$

If rank is n, $A^TA$ (nxn) is invertible

$$(A^TA)^{-1}A^T = A_{left}^{-1}$$

Because $(A^TA)^{-1}A^TA = I$

$A_{left}^{-1}$ is nxm, $A$ is mxn, and $I$ is nxn

$AA^T$ is bad in this case, it is not invertible

If multiply left inverse on the right, what is it?

$$AA_{left}^{-1} = A(A^TA)^{-1}A^T = P$$

It is a projection matrix into a column space


## Right Inverse

Full row rank r = m < n. Rows are independent but columns are not

$N(A^T) = 0$ What dimension in nullspace of A transpose? How many free variables? n - m free variables

Infinitely many solution to $Ax = b$ 

$$AA^{-1} = I$$

$$AA^T(AA^T)^{-1} = I$$

$$A^T(AA^T)^{-1} = A_{right}^{-1}$$

If multiply right inverse on the left, what is it?

$$A_{right}^{-1}A = A^T(AA^T)^{-1}A = P$$

It is a projection matrix onto row space



## Pseudoinverse

There is a vector $\vec x$ in the row space, multiply by $A$, we get an output in the column space. $A\vec x$ is the combination in column space. 

Connection between $\vec x$ and $A\vec x$ is one-to-one. Both have same, let say r, dimension space. Matrix A kills $N(A)$ and pseudoinverse kills $N(A^T)$

If $\vec x \neq \vec y$ and both in row space, then $Ax \neq Ay$ in column space

The reverse is pseudoinverse. $y = A^{+}(Ay)$

Proof:

Suppose $Ax = Ay \implies A(x - y) = 0$. This means $x - y$ in the nullspace. Because $x$ and $y$ in the row space, so $x - y$ is also in the row space. $x - y$ is in the nullspace and row space at the same time, what is it? Zero vector


### Find the pseudoinverse $A^{+}$

Start from SVD: $A = U\Lambda V^T$

$$\Lambda = \begin{bmatrix}\sigma_1 & ... \\ 0 & \ddots \\ 0 & 0 & \sigma_r \\ 0&0&0&0 \end{bmatrix}$$

with m rows, n columns, r rank

$$\Lambda^{+} = \begin{bmatrix}1/\sigma_1 & ... \\ 0 & \ddots \\ 0 & 0 & 1/\sigma_r \\ 0&0&0&0 \end{bmatrix}$$

with n  rows, m columns, rank r

$$\Lambda \Lambda^{+} = \begin{bmatrix}1 & ... \\ 0 & \ddots \\ 0 & 0 & 1 \\ 0&0&0&0 \end{bmatrix}$$

Diagonal 1 and the rest is 0. mxn multiply nxm produces mxm. Projection onto column space

$$\Lambda^{+}\Lambda  = \begin{bmatrix}1 & ... \\ 0 & \ddots \\ 0 & 0 & 1 \\ 0&0&0&0 \end{bmatrix}$$

Diagonal 1 and the rest is 0. nxm multiply mxn produces nxn. Similar to the above but actuall different size. Projection onto row space

$A^{+} = V\Lambda^{+} U^T$


In statistics, sometimes taking many measurements make $A^TA$ singular, hence not invertible, this is when pseudoinverse becomes useful
