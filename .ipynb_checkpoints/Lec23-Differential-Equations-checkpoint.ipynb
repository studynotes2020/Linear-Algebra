# Differential Equations

$$\frac{du_1}{dt} = -u_1 + 2u_2$$

$$\frac{du_2}{dt} = u_1 - 2u_2$$

$$A = \begin{bmatrix} -1 & 2 \\ 1 & -2 \end{bmatrix}$$

Initial condition $u(0) = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$

The determinant of A is zero, so it is a singular matrix, that tell us $\lambda = 0$ and the trace is +3, so the another $\lambda = -3$

We can also compute
$$|A - \lambda I| = \begin{vmatrix} -1-\lambda & 2 \\ 1 & -2-\lambda \end{vmatrix}$$

$$\lambda^2 + 3\lambda = 0$$

$\lambda = 0, -3$

So the solution will have two parts, $e^{0t}, e^{-3t}$. The second part will disappear when time goes on, the first part is a constant, hence a steady state

Case eigenvalues 0: $x_1 = \begin{bmatrix} 2 \\ 1 \end{bmatrix}$, $Ax_1 = 0x_1$

Case eigenvalues -3: $x_2 = \begin{bmatrix} 1 \\ -1 \end{bmatrix}$, $Ax_2 = -3x_2$

Solution: $$u(t) = c_1 e^{\lambda_1 t} x_1 + c_2 e^{\lambda_2 t} x_2$$

They both are solutions to the differential equation. 

Check: $\frac{du}{dt} = Au$ by plugging in $u = e^{\lambda_1 t}x_1$

$$\lambda_1 e^{\lambda_1 t}x_1 = A e^{\lambda_1 t}x_1  \checkmark$$

Then we need to find $c_1, c_2$, by using the result of eigenvalues and eigenvectors

$$u(t) = c_1.1\begin{bmatrix} 2 \\ 1 \end{bmatrix} + c_2 e^{-3t} \begin{bmatrix} 1 \\ -1 \end{bmatrix}$$

From initial condition, t = 0

$$u(0) = \begin{bmatrix} 1 \\ 0 \end{bmatrix} = c_1 \begin{bmatrix} 2 \\ 1 \end{bmatrix} + c_2 \begin{bmatrix} 1 \\ -1 \end{bmatrix}$$

Solve the simulataneous equation, $c_1 = c_2 = \frac13$

$$u(t) = \frac13 \begin{bmatrix} 2 \\ 1 \end{bmatrix} + \frac13 e^{-3t} \begin{bmatrix} 1 \\ -1 \end{bmatrix}$$

When $t \to \infty$

$$u(\infty) = \frac13 \begin{bmatrix} 2 \\ 1 \end{bmatrix}$$


1. Stability is when $u(t) \to 0$ which means we need all $e^{\lambda t} \to 0$, implies $\lambda < 0$. But what if eigenvalues is complex? $e^{(-3+6i)t}$. Then we take its absolute value $|e^{(-3+6i)t}|$ which is equal to $e^{-3t} $ because $|e^{6it}| = 1$. So real part of $\lambda$ has to be less than zero


2. Steady state $\lambda_1 = 0$ and other $Re(\lambda) < 0$


3. Blow up if any $Re(\lambda) > 0$

## Decouple matrix

$$\frac{du}{dt} = Au$$

The matrix $A$ couple $u$ and $u'$. To uncouple (means diagonalizing it), set $u = Sv$. $S$ is eigenvector matrix

$$S\frac{dv}{dt} = ASv$$

$$\frac{dv}{dt} = S^{-1}ASv = \Lambda v$$

$$\frac{dv_1}{dt} = \lambda_1 v_1 \quad ... \quad \frac{dv_n}{dt} = \lambda_n v_n$$

It is a system of equations but they are not connected

Solution:
$$v(t) = e^{\Lambda t}v(0)$$

$$u(t) = Se^{\Lambda t}S^{-1}u(0) = e^{At}u(0)$$

Define $$ e^{At} = Se^{\Lambda t}S^{-1}$$

What is $ e^{At}$?


## Matrix Exponential $e^{At}$

Use the power series to define

$$e^{At} = I + At + \frac{(At)^2}{2} + \frac{(At)^3}{6} + ... + \frac{(At)^n}{n!}$$

This is from Taylor series where $e^x = \sum\frac{x^n}{n!}$

The other famous Taylor series is $\frac{1}{1 - x} = \sum_0^\infty x^n$, which can also put in matrix form

$$(I - At)^{-1} = I + At + (At)^2 + (At)^3 + ...$$

We can do same thing to matrix just like we do for functions. For the second series, eigenvalues of $At$ should be less than 1 for the series to converge.

We are interested in the first series in this lecture.

Because $A = S\Lambda S^{-1}$

$$e^{At} = I + S\Lambda S^{-1}t + \frac{S\Lambda^2 S^{-1}}{2}t^2 + ... + \frac{S\Lambda^n S^{-1}}{n!}t^n$$

We can see the first term $I = SS^{-1}$, then factor out $SS^{-1}$

$$e^{At} = S e^{\Lambda t}S^{-1}$$

This equation doesn't always work. There is an assumption that $A$ can be diagonalized.



## What is $e^{\Lambda t}$

$$\Lambda = \begin{bmatrix} \lambda_1 & 0 & 0 \\ 0 & \ddots & 0 \\ 0 & 0 & \lambda_n \end{bmatrix}$$

$$e^{\Lambda t} = \begin{bmatrix} e^{\lambda_1 t} & 0 & 0 \\ 0 & \ddots & 0 \\ 0 & 0 & e^{\lambda_n t} \end{bmatrix}$$

To be stable, every real part of eigenvalues less than 0 (negative)


## Graph Imaginary vs Real

For the differential equations to be stable, the $e^{At} \to 0$, which is the region in the left half of the plane.

For the power of matrix A to go to zero, $A^n \to 0$ the eigenvalue is inside the unit circle, $|\lambda| < 1$


## Convert higher order differential equation to lower order

$$y'' + by' + ky = 0$$

How to change 1 2nd order equation into 2x2 1st order equation?

Take $u = \begin{bmatrix}y' \\ y \end{bmatrix}$, then 

$$u' = \begin{bmatrix}y'' \\ y' \end{bmatrix} = \begin{bmatrix} -b & k \\ 1 & 0\end{bmatrix} \begin{bmatrix} y' \\ y \end{bmatrix}$$

We can change 5th order to 5x5 1st order. 5th order equation in the first row, then four trivial equations beneath

$$\begin{bmatrix} c_1 & c_ 2 & c_3 & c_4 & c_5 \\ 1 \\ 0 & 1 \\ 0 & 0 & 1 \\ 0 & 0 & 0 & 1  \end{bmatrix}$$
