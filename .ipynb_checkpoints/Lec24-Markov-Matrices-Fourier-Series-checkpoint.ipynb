# Markov Matrices

$$A = \begin{bmatrix}.1 & .01 & .3 \\ .2 & .99 & .3 \\ .7 & 0 & .4 \end{bmatrix}$$

Two properties for a Markov matrix

1. All entries >= 0 (probability never negative)

2. All columns add to 1

Steady state when $\lambda = 1$

Key points

1. $\lambda = 1$ is an eigenvalue and the associated eigenvector $\vec x_1 \geq 0$. No negative component

2. All other $|\lambda_i| < 1$

$$u_k = A^k u_0 = c_1\lambda_1^k x_1 + c_2\lambda_2^k x_2 + ...$$

If $\lambda_1 = 1$ and other lambdas small than 1, those other lambdas goes to zero when time goes on and $u_k \to c_1x_1$. So $x_1$ is part of $u_0$ 

Case eigenvalue is 1:

$$A - I = \begin{bmatrix}-.9 & .01 & .3 \\ .2 & -.01 & .3 \\ .7 & 0 & -.6 \end{bmatrix}$$

This matrix is singular. Eigenvalue is the value of shifting A so that it become singular

All columns of $A-I$ add to zero implies $A-I$ is singular. Why?

For a matrix to be singular, it means that the columns are dependent. The determinant is zero when the columns are dependent. The columns add to zero means they are dependent 

It is easier to see adding all the rows become zero. Because (1,1,1) is in $N(A^T)$, so the rows are dependent

What combination of column make zero?

It is the eigenvector $x_1$, it is in $N(A)$

Another point, eigenvalue of $A$ and eigenvalue of $A^T$ is the same! Why?

The determinant of a matrix and the determinant of its transpose are the same. 

$$det(A - \lambda I) = det(A - \lambda I)^T$$

Transpose $A$ is $A^T$, transpose $\lambda I$ is $\lambda I$

$$(A - \lambda I)^T = A^T - \lambda I$$

$$det(A^T - \lambda I) = det(A - \lambda I) = 0 $$

However the eigenvector is different. Because 

$$\begin{bmatrix}-.9 & .01 & .3 \\ .2 & -.01 & .3 \\ .7 & 0 & -.6 \end{bmatrix} \begin{bmatrix}.6 \\ 33 \\ .7 \end{bmatrix} = 0$$ 

$$\vec x_1 = \begin{bmatrix}.6 \\ 33 \\ .7 \end{bmatrix}$$

# Example

$$u_{k+1} = Au_k$$

A is Markov. 


$$\begin{bmatrix}u_{cal} \\ u_{mass} \end{bmatrix}_{k+1} = \begin{bmatrix}.9 & .2 \\ .1 & .8  \end{bmatrix} \begin{bmatrix}u_{cal} \\ u_{mass} \end{bmatrix}_k$$

Let say the process starts with 

$$\begin{bmatrix}u_{cal} \\ u_{mass} \end{bmatrix}_0 = \begin{bmatrix}0 \\ 1000 \end{bmatrix}, \quad \begin{bmatrix}u_{cal} \\ u_{mass} \end{bmatrix}_1 = \begin{bmatrix}200 \\ 800 \end{bmatrix}$$

Proceed for many steps what is the result? To answer the question, find eigenvalue and eigenvector

Eigenvalues, $\lambda = 1, .7$

Case $\lambda = 1$, $v_1 = \begin{bmatrix}2 \\ 1 \end{bmatrix} $

What is the population at infinity? eigenvector gives us the steady state. The population should add up to 1000, so 

$$\begin{bmatrix}u_{cal} \\ u_{mass} \end{bmatrix}_\infty = 1000 \begin{bmatrix} \frac23 \\ \frac13 \end{bmatrix}$$

What if finite 100 step? then we need find second eigenvector

Case $\lambda = .7$, $v_2 = \begin{bmatrix}-1 \\ 1 \end{bmatrix} $

$$u_k = c_1 1^k \begin{bmatrix}2 \\ 1 \end{bmatrix} + c_2 .(7)^k \begin{bmatrix}-1 \\ 1 \end{bmatrix}$$

Plug in initial condition

$$u_0 = \begin{bmatrix}0 \\ 1000 \end{bmatrix} = c_1 \begin{bmatrix}2 \\ 1 \end{bmatrix} + c_2 \begin{bmatrix}-1 \\ 1 \end{bmatrix}$$

Solving the equations, we get

$c_1 = \frac{1000}{3}, c_2 = \frac{2000}{3}$

## Projections with orthonormal basis $q_1, ..., q_n$

Any v can be expanded (projected) onto a basis

$$v = x_1 q_1 + x_2 q_2 + ... + x_n q_n$$

Can we get a formula for $x_1$? Yes, take the inner product of its associated $q^T$ at both side

$$q_1^Tv = x_1 q_1^T q_1 + 0 + 0 + ...$$

Because any basis inner product with other basis is zero and inner product with itself is 1

$$x_1 = q_1^Tv$$

Writing in matrix form

$$\begin{bmatrix}\vec q_1 & ... \vec q_n \end{bmatrix} \begin{bmatrix}x_1 \\ \vdots \\ x_n \end{bmatrix} = v$$

$$Qx = v$$

$$x = Q^{-1}v = Q^T v$$

Q is orthonormal so its inverse is equal to its transpose. Hence we get the same result

$$x_1 = q_1^T v$$



## Fourier Series

Fourier series built on the above result.

$$f(x) = a_0 + a_1 \cos x + b_1 \sin x + a_2 \cos 2x + b_2 \sin 2x + ...$$

The difference is this is in infinite dimensional spaces compare to finite dimensional spaces above. The basis is functions, instead of vectors.

Inner product of the vectors

$$v^Tw = v_1w_1 + ... + v_nw_n$$

Inner product of the functions

$$f^Tg = \int_0^{2\pi} f(x)g(x)\ dx$$

Analog of addition in continuum is integral and those cosines and sines functions have period 2pi, so we better integrate between that range

Periodic function: $$f(x) = f(x + 2\pi)$$

Orthogonality: $$\int_0^{2\pi} \sin x \cos x \ dx = \frac12 (\sin x)^2 |_0^{2\pi} = 0$$


How to get $a_1$

Take inner product of everything with $\cos x$ and integrate

On RHS
$$\int_0^{2\pi} f(x) \cos x \ dx$$

On LHS, all the terms become zero accept the term associate with $a_1$

$$a_1 \int_0^{2\pi} (\cos x)^2 \ dx = a_1 \pi$$

$$a_1 = \frac{1}{\pi} \int_0^{2\pi} f(x) \cos x \ dx$$

So, Fourier series is an expansion in an orthonormal basis
