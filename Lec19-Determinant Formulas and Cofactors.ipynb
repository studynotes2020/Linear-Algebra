## Determinant Formula

Property 1:
$\begin{vmatrix} 1 & 0 \\ 0 & 1 \end{vmatrix} = 1$

Property 2:
$\begin{vmatrix} 0 & 1 \\ 1 & 0 \end{vmatrix} = -1$

Property 3:
$\begin{vmatrix} a & b \\ c & d \end{vmatrix} =
\begin{vmatrix} a & 0 \\ c & d \end{vmatrix} + 
\begin{vmatrix} 0 & b \\ c & d \end{vmatrix} = 
\begin{vmatrix} a & 0 \\ c & 0 \end{vmatrix} + 
\begin{vmatrix} a & 0 \\ 0 & d \end{vmatrix} +
\begin{vmatrix} 0 & b \\ c & 0 \end{vmatrix} + 
\begin{vmatrix} 0 & b \\ 0 & d \end{vmatrix} $

Those have column of zero become zero

$\begin{vmatrix} a & b \\ c & d \end{vmatrix} =
\begin{vmatrix} a & 0 \\ 0 & d \end{vmatrix} +
\begin{vmatrix} 0 & b \\ c & 0 \end{vmatrix}  = 
a.d\begin{vmatrix} 1 & 0 \\ 0 & 1 \end{vmatrix} +
(-b.c)\begin{vmatrix} 1 & 0 \\ 0 & 1 \end{vmatrix} =
a.d - b.c$

By using this method, we can also get the formulas for 3x3, nxn.

### Big Formula

$$det A = \sum_{n!} \pm a_{1\alpha} a_{2\beta}a_{3\gamma}a_{n\omega}$$

The first row can be chosen $n!$ ways, second row chosen $(n-1)!$, ... last row chosen 1 way, so $n!$

$(\alpha, \beta, \gamma, ...,\omega) = permutation (1,2,...,n)$

The first row can be chosen n ways, second row chosen n-1, ... last row chosen 1 way, so $n!$

### Cofactor formula

$det A = a_{11}C_{11} + a_{12}C_{12} + ... + a_{1n}C_{1n}$
