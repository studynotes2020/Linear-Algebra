# Complex matrix

Each entry is a complex number, so the matrix in n dimensional complex space $C^n$

$$z = \begin{bmatrix}z_1 \\ z_2 \\ \vdots \\ z_n \end{bmatrix}$$

$z^T z$ no good, doesn't give length square. 

What we want is $\overline z^T z$ gives positive length square. Take z with length $\sqrt{2}$

$$\overline z^T z = \begin{bmatrix}1 & -i \end{bmatrix} \begin{bmatrix}1 \\ i \end{bmatrix} = \begin{bmatrix} 2 \end{bmatrix}$$

Define $z^H := \overline z^T$. z Hermitian

$$z^H z = |z_1|^2 + ... + |z_m|^2$$

For real matrix, inner product is $\overline y^T x$

For complex matrix, inner product is $y^H x$

Symmetric: $A^T = A$ no good if $A$ is complex

Instead, symmetric for complex matrix is $\overline A^T = A$

Symmetric complex matrix is called Hermitian matrix. Example

$$A^H = A = \begin{bmatrix}2 & 3 + i \\ 3 - i & 5 \end{bmatrix}$$


## Unitary matrix

A square nxn matrix, orthonormal columns (perpendicualr and unit vectors)

$q_1, q_2, ..., q_n$

If perpendicular

$$q_i^H q_j = \begin{cases} 0 &i \neq j\\
 1 &i = j\end{cases}$$
 
 $$Q^H Q = I$$ 

# Fast Fourier Transform

$$F_n = \begin{bmatrix}1 & 1 & 1 & ... & 1\\ 1 & w & w^2 & ... & w^{n-1}\\ 1 & w^2 & w^4 & ... & w^{2(n-1)}\\ \vdots \\ 1 & w^{n-1} & w^{2(n-1)} & ... & w^{(n-1)^2}\end{bmatrix}$$

Entry $$(F_n)_{ij} = w^{ij}$$

where $i, j = 0, ..., n - 1$

Definition of w

$$w^n = 1 \implies w = e^{i2\pi/n} = \cos \frac{2\pi}{n} + i\sin \frac{2\pi}{n}$$

It is easier to work with exponential form

If n = 6, $w = e^{i2\pi/6}$. The angle is $2\pi/6$ and length always 1

If n = 4, $w^4 = 1 \quad w = e^{i2\pi/4} = i$

$$F_4 = \begin{bmatrix}1 & 1 & 1 & 1\\ 1 & i & i^2 & i^3 \\ 1 & i^2 & i^4 & i^6 \\ 1 & i^3 & i^6 & i^9 \end{bmatrix} = \frac12 \begin{bmatrix}1 & 1 & 1 & 1\\ 1 & i & -1 & -i \\ 1 & -1 & 1 & -1 \\ 1 & -i & -1 & i \end{bmatrix}$$

The columns are orthogonal. Take any inner product of two columns, $q_i^H q_j$  it is zero

$\frac12$ in front to make it orthonormal. Then we can do $F^{-1}$ quickly

$$F_4^H F_4 = I \implies F_4^{-1} = F_4^H$$

There is connection between $F_n$ and $F_{n/2}$

$$e^{2 . 2\pi/64} = e^{i2\pi/32} \implies (w_{64})^2 = w_{32}$$

$$F_{64} =  \begin{bmatrix}I & D\\ I & -D  \end{bmatrix} \begin{bmatrix}F_{32} & 0\\ 0 & F_{32}  \end{bmatrix} P $$

$64^2$ operation reduces to $2(32)^2 + 32$ operation

$P$ is permutation matrix

$D = \begin{bmatrix}1 & \\  & w \\ & & w^2 \\ & & & \ddots \\ & & & & w^{31}\end{bmatrix} $

Each $F_{32}$ then can be broken down to $F_{16}$

$2(32)^2 + 32$ operation reduces to $2[2(16)^2 + 16]^2 + 32$ operation. Do this recursively, until 1.

Finally, 6x32 which is $log_2 64$. 

$O(n^2)$ reduces to $\frac12 n log_2 n$

FFT for nxn matrix, does it not in $n^2$ square steps but $\frac12 n log_2 n$ steps
