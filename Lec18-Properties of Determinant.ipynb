## Properties 1, 2, 3, ..., 10

determinant A = $\begin{vmatrix} A \end{vmatrix}$

### Property 1

det $I$ = 1

$\begin{vmatrix} 1 & 0 \\ 0 & 1 \end{vmatrix} = 1$

### Property 2

What happen to determinant when exchanging rows: Reverse sign of det

det $P$ = $\begin{cases}-1 & \text{if number of exchange is odd} \\ 
1 & \text{if number of exchange is even} \end{cases}$

$\begin{vmatrix} 0 & 1 \\ 1 & 0 \end{vmatrix} = -1$

### Property 3a

$\begin{vmatrix} ta & tb \\ c & d \end{vmatrix} = t\begin{vmatrix} a & b \\ c & d \end{vmatrix}$

### Property 3b

Linear on each row

$\begin{vmatrix} a + a' & b + b' \\ c & d \end{vmatrix} = 
\begin{vmatrix} a & b \\ c & d \end{vmatrix} + 
\begin{vmatrix} a' & b' \\ c & d \end{vmatrix}$

Not saying $det(A + B) = det A + det B$


### Property 4

2 equal rows $\to$ det = 0

Reasoning: exchange those equal rows results in the same matrix, so determinant no change. But property 2 says the sign should change. Hence the only possibility is the determinant is zero

### Property 5

Can substract l x rowi from row k, the determinant doesn't change

$det U = det A$

$\begin{vmatrix} a & b  \\ c-la & d-lb \end{vmatrix} = 
\begin{vmatrix} a & b \\ c & d \end{vmatrix} + 
\begin{vmatrix} a & b \\ -la & -lb \end{vmatrix}$ using property 3b

$ = \begin{vmatrix} a & b \\ c & d \end{vmatrix} - l\begin{vmatrix} a & b \\ a & b \end{vmatrix}$ using property 3a

$ = \begin{vmatrix} a & b \\ c & d \end{vmatrix}$ using property 4

### Property 6

Row of zero $\to$ det A = 0

$ 0 \begin{vmatrix} a & b \\ c & d \end{vmatrix}
= \begin{vmatrix} 0.a & 0.b \\ c & d \end{vmatrix}
= 0$

### Property 7

$det U = \begin{vmatrix} d_1 & * & * \\ 0 & d_2 & * \\ 0 & 0 & d_3 \end{vmatrix}
= d_1.d_2.d_3...d_n$ (product of pivots)

Prove using property 3a

We can make every off-diagonal entries to be zero by using property 5

$\begin{vmatrix} d_1 & * & * \\ 0 & d_2 & * \\ 0 & 0 & d_3 \end{vmatrix} = 
\begin{vmatrix} d_1 & 0 & 0 \\ 0 & d_2 & 0 \\ 0 & 0 & d_3 \end{vmatrix}
 = d_3.d_2.d_1 \begin{vmatrix} 1 & 0 & 0 \\ 0 & 1 & * \\ 0 & 0 & 1 \end{vmatrix}
 = d_n...d_3.d_2.d_1$ using property 1

### Property 8

det A = 0 when A is singular. If singular, elimination get row of zeros

det A $\neq$ 0 when A is invertible. If invertible, elimination get $U \to D \to d_1.d_2...d_n$ 

$ \begin{vmatrix} a & b \\ c & d \end{vmatrix}
\to \begin{vmatrix} a & b \\ 0 & d - cb/a \end{vmatrix} = ad - bc$

### Property 9

$det(AB) = det(A) . det(B)$

but $det(A + B) \neq det(A) + det(B)$

Using this property, we can get $det A^{-1}$

$A^{-1}A = I \to (det A^{-1})(det A) = 1$

$det A^{-1} = \frac1{det A}$

Also other facts from this property:

$det A^2 = (det A)^2$

$det 2A = 2^n det A$ bcs we are factoring out 2 from every row, there are n rows

### Property 10

$det A^T = det A$

$ \begin{vmatrix} a & b \\ c & d \end{vmatrix}
= \begin{vmatrix} a & c \\ b & d \end{vmatrix}$

If column zero, transpose column to row, so determinant is zero

Prove: $\begin{vmatrix} A^T \end{vmatrix} = \begin{vmatrix} A \end{vmatrix}$

$\begin{vmatrix} U^T L^T \end{vmatrix} = \begin{vmatrix} LU \end{vmatrix}$

$\begin{vmatrix} U^T \end{vmatrix} \begin{vmatrix}L^T \end{vmatrix} = \begin{vmatrix} L \end{vmatrix} \begin{vmatrix} U \end{vmatrix}$

Both sides are triangular matrices, so determinant is product of pivots
