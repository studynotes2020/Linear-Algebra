# Symmetric Matrices

A symmetric matrix means $$A = A^T$$

Two key facts

1. The eigenvalues are REAL

2. The eigenvectors are perpendicular

Identity is a symmetric matrix. Every vector is its eigenvectors. Is all vectors perpendicular to identity? No. Actually, the second fact means the eigenvectors can be chosen perpendicular

Usually a matrix can be decomposed as eigenvector matrix and eigenvalue matrix

$$A = S\Lambda S^{-1}$$

For a symmetric matrix, we can choose orthonormal eigenvectors, so

$$A = Q\Lambda Q^{-1}$$

For orthonormal matrix, $Q^{-1} = Q^T$

$$A = Q\Lambda Q^T$$

This is called Spectral theorem.

Why real eigenvalues?

$$Ax = \lambda x \implies A \overline x = \overline \lambda \overline x$$

At this point, $\lambda, x$ can be complex. It means their conjugate also equal. But we are only talking about real $A$ and if $A$ is real, $\lambda, x$ cannot be complex

Taking transpose of both side, and use the fact that for a symmetric matrix $A = A^T$

$$ \implies \overline x^T A^T = \overline x^T \overline \lambda \implies \overline x^T A = \overline x^T \overline \lambda \implies \overline x^T Ax = \overline x^T \overline \lambda x$$

Back to this equation $A x =  \lambda  x$

Multiply both side with $\overline x^T$ 

$$\overline x^T A x = \overline x^T \lambda x$$

We can see that they have both the same LHS, $\overline x^T A x$, so RHS must be equal

$$\overline x^T \overline \lambda x = \overline x^T \lambda x$$

Because $\lambda$ is a number, it can be brought in front

$$\overline \lambda \overline x^T  x = \lambda \overline x^T  x$$

$$\overline \lambda = \lambda$$

So $\lambda$ is real because the number is equal to its conjugate only when it is real

$$x^Tx = \begin{bmatrix}\overline x_1 & \overline x_2 & ... & \overline x_n \end{bmatrix} \begin{bmatrix}x_1 \\ x_2 \\ \vdots \\ x_n \end{bmatrix}$$

$$\overline x_1 x_1 = (a-bi)(a+bi) = a^2 + b^2$$

So $ \overline x^T  x$ can be safely cancelled out and reach the conclusion

## Good matrices

Good means real eigenvalues and perpendicular eigenvectors which implies $A = A^T$. In most of the applications, $A$ is real.

If $A$ is complex, it is good if $A = \overline A^T$

In short, if $A = A^T$

$$A = Q\Lambda Q^T$$

$$\begin{bmatrix}\vec q_1 & \vec q_2 & ... \end{bmatrix} 
\begin{bmatrix}\lambda_1 & 0 & 0 \\ 0 & \lambda_2 & 0 \\ 0 & 0 & \ddots \end{bmatrix}
\begin{bmatrix}q_1^T \\ q_2^T \\ \vdots \end{bmatrix}
= \lambda_1 \vec q_1 \vec q_1^T + \lambda_2 \vec q_2 \vec q_2^T + ...$$

Every symmetric matrix is a combination of perpendicular projection matrices

For a very big matrix like 50x50, it is not feasible to find its eigenvalues by taking its determinant and solve for the polynomial roots

But Matlab can compute pivots quickly. Signs of pivots are the same as signs of eigenvalues. Number of positive pivots is equal to the number of positive eigenvalues. 

It then shift the matrix by 7I, then take the pivots of that matrix, and you would know how many eigenvalues of the original were above seven or below seven.

Product of the pivots is equal to the product of the eigenvalues because they both equal the determinant. 

Somehow it uses these facts to find all eigenvalues

## Positive Definite Symmetric Matrix

1. All eigenvalues are positive

2. All pivots are positive

3. All sub-determinants are positive

For a matrix
$$\begin{bmatrix}5 & 2 \\ 2 & 3 \end{bmatrix}$$

pivots are 5, 11/5, both positive

$$\lambda^2 - 8\lambda + 11 = 0$$

$\lambda = 4 \pm \sqrt{5}$ are also both positive

We want this matrix in differential equation because we know the sign of eigenvalues, we know it is stable or not

The determinant is positive too, but this is not enough. For example, this matrix has positive determinant too

$$\begin{bmatrix}-1 & 0 \\ 0 & -3 \end{bmatrix}$$

So, the test should be - All sub-determinants are positive

The above matrix fails the test because -1 is negative
