# Positive Definite Matrix

How do I know if this matrix if positive definite matrix? and why is this important?

$$A = \begin{bmatrix} a & b \\ b & c \end{bmatrix}$$

Test. Each test is a complete test

1. Eigenvalue test 
$\lambda_1 > 0, \lambda_2 > 0$


2. Determinant test
$a > 0, ac - b^2 > 0$


3. Pivots test 
$a > 0,  \frac{ac - b^2}{a} > 0$


4. This is the definition of positive definite matrix
$x^T Ax > 0$


### Examples

$$A = \begin{bmatrix} 2 & 6 \\ 6 & d \end{bmatrix}$$

If d = 19, it is positive definite. 

If d = 18, it is positive semidefinite. Then $\lambda = 0, 20$, It is the zero that causes it semidefinite. Pivots is 2 only and doesn't pass pivot test

$$x^TAx = \begin{bmatrix} x_1 & x_2 \end{bmatrix} \begin{bmatrix} 2 & 6 \\ 6 & 18 \end{bmatrix} \begin{bmatrix} x_1\\ x_2 \end{bmatrix}
= \begin{bmatrix} 2x_1^2 + 12x_1x_2 + 18x_2^2 \end{bmatrix}$$

The result is the form $ax^2 + 2bxy + cy^2$. It becomes quadratic form (not linear). If $ax^2 + 2bxy + cy^2 > 0$ for any x and y, then A is a positive definite matrix

If d = 7, then $$A = \begin{bmatrix} 2 & 6 \\ 6 & 7 \end{bmatrix}$$

There is some $x_1, x_2$ that make $2x_1^2 + 12x_1x_2 + 7x_2^2$ negative

Take $x_1 = 1, x_2 = -1$, then $2 - 12 + 7 = -3 < 0$

### $2x^2 + 12xy + 20y^2$

In Calculus, for a minimum, its first derivative is zero, and second derivative is positive

For a function, $f(x_1, x_2, ..., x_n)$. The minimum will be when the matrix of 2nd derivative is positive definite. Positive for a number will be translated into positive definite for a matrix

Let completing the square for

$2x^2 + 12xy + 20y^2 = 2(x + 3y)^2 + 2y^2$ 

If we cut the graph at height = 1, that means $2x^2 + 12xy + 20y^2 = 1$ which is an ellipse curve

Those coefficients 2, 3, 2 actually comes from Gaussian elimination. Completing the square is the old method of Gaussian elimination

pivot is 2, multiplier is 3 

$$\begin{bmatrix}2 & 6 \\ 6 & 20 \end{bmatrix} = \begin{bmatrix}1 & 0 \\ 3 & 1 \end{bmatrix} \begin{bmatrix} 2 & 6 \\ 0 & 2\end{bmatrix} $$

$$A = LU$$

So completing the square is elimination

Pivots are what go outside the square (2, 2). So positive pivots give positive definiteness

Matrix of 2nd derivative for 2 dimensional
$$\begin{bmatrix}f_{xx} & f_{xy} \\ f_{yx} & f_{yy} \end{bmatrix} $$

For a function of two variables, a minimum, implies first derivative is zero, and matrix of 2nd derivative is positive definite or determinant is positive


### 3x3 Example

Is this matrix positive definite?

$$A = \begin{bmatrix}2 & -1 & 0 \\-1 & 2 & -1 \\ 0 & -1 & 2 \end{bmatrix}$$

Its determinants are 2, 3, 4

Pivots are 2, 3/2, 4/3

Eigenvalues are $2-\sqrt{2}, 2, 2+\sqrt{2}$

$$x^T Ax = 2x_1^2 + 2x_2^2 + 2x_3^2 - 2x_1x_2 - 2x_2x_3 > 0$$

The graph is a bowl

If completing the square, those three squares will be multiplied by the pivots, and the pivots are all positive. So squares times positive pivots will be a graph of bowl

Suppose cutting the bowl at height = 1

$2x_1^2 + 2x_2^2 + 2x_3^2 - 2x_1x_2 - 2x_2x_3 = 1$. It is an ellipsoid or rugby ball

This case, three eigenvectors are different. It has major, minor and middle axis, and those three axis will be in the direction of eigenvectors. Length of those axis determined by the eigenvalues

From Principal Axis Theorem, we can decompose A into eigenvector and eigenvalue matrix

$$A = Q\Lambda Q^T$$
