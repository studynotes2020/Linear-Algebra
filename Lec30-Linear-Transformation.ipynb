# Linear Transformation

Many physicists don't like coordinates, they want to know what happen to all space

without coordinates: matrix

with coordinates -> Matrix

### Example 1: Projection

We can do projection without coordinates

$T: R^2 \to R^2$

Transformation is a map between input and output. 

Rules of Linear Transformation

1. $T(v + w) = T(v) + T(w)$

2. $T(cv) = cT(v)$

So, $T(cv + dw) = cT(v) + dT(w)$


### Example 2: Shift whole plane

Every vector $\vec v$, shift them by $v_0$. Is this transformation linear?

No. If double the length of $\vec v$, the transformation still goes $v_0$, not double of $v_0$

In fact, $T(0) = 0$ is a linear transformation

$T(v) = ||v||, T: R^3 \to R$

It is true that zero vector goes to zero. Also double v, double length. But  v multiply by minus 2, the length does not multiply by minus 2


### Example 3: Rotation by $45^o$

$T: R^2 \to R^2$

$T(v)$ rotate $v$ by $45^o$

We can describe this without coordinates. If add then rotate, transformation also add and rotate. If double length, transformation also double length


### Example 4: Matrix A

$T(v) = Av$

This is a linear transformation

$A(v + w) = Av + Aw \checkmark$

$A(cv) = cAv \checkmark$


$$A = \begin{bmatrix}1 & 0 \\ 0 & -1 \end{bmatrix}$$

x stays the same, but y reverses. The house becomes upside down 


Start: $T: R^3 \to R^2$

Example: $T(v) = Av$

Input in $v \in R^3$, output $T(v) \in R^2$. so matrix $A$ is 2x3


Information needed to know transformation $T(v)$ for all inputs

$T(v_1), T(v_2), ..., T(v_m)$ for any input basis $v_1, ..., v_n$

If I know the input basis, and what transformation does to those basis, then I know transformation for all v. Because every v is some combination of those basis, $\vec v = c_1v_1 + ... + c_nv_n$

We know RHS, then we know LHS

$T(\vec v) =  c_1T(v_1) + ... + c_nT(v_n)$

## Coordinates come from a basis

Coordinate of $v = c_1v_1 + ... + c_nv_n$

If I change the basis, I change the coordinate. We have always been assuming we are working with standard basis.

$v = \begin{bmatrix}3 \\ 2 \\ 4 \end{bmatrix} = 3 \begin{bmatrix}1 \\ 0 \\ 0 \end{bmatrix} + 2 \begin{bmatrix}0 \\ 1 \\ 0 \end{bmatrix} + 4 \begin{bmatrix}0 \\ 0 \\ 1 \end{bmatrix}$

Actually, we could have chosen other basis. For example, we can use eigenvector basis




## Construct matrix A that represents linear transformation

$T: R^n \to R^m$

T transform from n dimension to m dimension

We need input basis and output basis

Choose basis $v_1, ..., v_n$ for input $R^n$

Choose basis $w_1, ..., w_m$ for output $R^m$

Take a vector v, express in input basis, multiply by right matrix A, which gives transformation in output basis

Suppose we want to project every vector in the plane into a line. We choose unit vector in the direction of the line and unit vector perpendicular to the line as output basis

Input $\vec v = c_1v_1 + c_2v_2$

Output $T(\vec v) = c_1v_1$

In other words, the coordinate of the input is $(c_1, c_2)$, transformed into $(c_1, 0)$

The matrix that do this transformation is
$$A = \begin{bmatrix}1 & 0 \\ 0 & 0 \end{bmatrix}$$

$$\begin{bmatrix}1 & 0 \\ 0 & 0 \end{bmatrix} \begin{bmatrix}c_1 \\ c_2 \end{bmatrix} = \begin{bmatrix}c_1 \\ 0 \end{bmatrix}$$\

Eigenvector basis is the good basis because it leads to diagonal matrix $\Lambda$. In this example, two eigenvectors are line in direction and prependicular to the input basis. Eigenvalues are 1 and 0


### Project onto $45^o$ line

use standard basis $v_1 = \begin{bmatrix}1 \\ 0 \end{bmatrix} = w_1$, $v_2 = \begin{bmatrix}0 \\ 1 \end{bmatrix} = w_2$

Projection matrix

$$P = \frac{aa^T}{a^Ta} = \begin{bmatrix}1/2 & 1/2 \\ 1/2 & 1/2 \end{bmatrix}$$


## Rule to find $A$

Given input bases (v1..vn) and output bases (w1 ... wm)

1st column of $A$: Apply $T(v_1) = a_{11}w_1 + a_{21}w_2 + ... + a_{m1}w_m$

2nd column of $A$: Apply $T(v_2) = a_{12}w_1 + ... + a_{m2}w_m$

A(input coordinate) = (output coordinate)


## Transformation for derivative

Input: $c_1 + c_2x + c_3x^2$, basis is $1, x, x^2$

Output (derivative): $c_2 + 2c_3x$, basis is $1, x$ 

So this is a transformation of 3 dimension space of input into 2 dimension space of output by derivative

$$A\begin{bmatrix}c_1 \\ c_2 \\ c_3 \end{bmatrix} = \begin{bmatrix}c_2 \\ 2c_3 \end{bmatrix}$$

By inspection, we can write 

$$A = \begin{bmatrix}0 & 1 & 0 \\ 0 & 0 & 2 \end{bmatrix}$$

Matrix multiplication really comes from linear transformation
