Positive definite of a matrix is analogy to a positive number

Positive definite means $x^TAx > 0$ except for x = 0. But the direct way to test it is with pivots, eigenvalues and determinant

But where does positive definite matrix comes from?

They come from least square

Suppose a symmetry positive definite matrix, its inverse is also positive definite. 1 over eigenvalue is also positive.

Suppose A and B are positive definite, how about A + B?

$x^T(A + B)x > 0$ because $x^TAx > 0$ and $x^TBx > 0$

Now A is mxn. $A^TA$ is square and symmetric, is it positive definite?

We find $x^TA^TAx \implies (Ax)^T (Ax) \implies |Ax|^2 \geq 0$ 

If A has rank = n (n independent columns) then A is positive definite

Also, one great thing about positive definite matrix is that the pivot position will not be a small numbers or zeroes, so no row exchange needed. They are great matrices to compute. 


# Similar Matrices

A and B (both are nxn) are similar means for some matrix M

$$B = M^{-1}AM$$

Example: A is similar to $\Lambda$ (eigenvalue matrix)

$$S^{-1}AS = \Lambda$$

Suppose  
$$A = \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix}
\quad  \Lambda = \begin{bmatrix} 3 & 0 \\ 0 & 1 \end{bmatrix}$$

Take random matrix M
$$M = \begin{bmatrix} 1 & -4 \\ 0 & 1 \end{bmatrix}$$

$$M^{-1}AM = \begin{bmatrix} 1 & -4 \\ 0 & 1 \end{bmatrix}
\begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix}
\begin{bmatrix} 1 & 4 \\ 0 & 1 \end{bmatrix} = 
\begin{bmatrix} 1 & -4 \\ 0 & 1 \end{bmatrix}
\begin{bmatrix} 2 & 9 \\ 1 & 9 \end{bmatrix} = 
\begin{bmatrix} -2 & -15 \\ 1 & 6 \end{bmatrix} = B$$

We say $A, B, \Lambda$ is similar matrix. They all have the same eigenvalues $\lambda = 3, 1$

Similar matrices have the same eigenvalues!! (same number of eigenvectors)

Another matrix that has same eigenvalues (just make it triangular) 

$$\begin{bmatrix} 3 & 7 \\ 0 & 1 \end{bmatrix}
\quad \begin{bmatrix} 1 & 7 \\ 0 & 3 \end{bmatrix}$$

Why are they all share the same eigenvalues?

Starts with

$Ax = \lambda x$, try to get $B = M^{-1}AM$ into picture

$$M^{-1}AMM^{-1}x = \lambda M^{-1}x$$

$$B (M^{-1}x) = \lambda (M^{-1}x)$$

So we know $B$ has the same eigenvalues, but eigenvectors are not the same

The eigenvector of $B = M^{-1}x$ where $x$ is eigenvector of $A$

# Jordan Theorem

Every square matrix A is similar to a Jordan matrix J

$$J = \begin{bmatrix} J_1 & 1 & 0 & 0 \\ 0 & J_2 & 1 & 0 \\ \vdots & \vdots & \ddots & 1 \\ 0 & 0 & 0 & J_d \end{bmatrix}$$

Number of blocks is equal to the number of eigenvectors

Good case: $J = \Lambda$

Jordan also covered all cases by including cases of repeated eigenvalues and missing eigenvectors
